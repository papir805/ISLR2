### Part c) Logistic Regression Confusion Matrix
Compute the confusion matrix and overall fraction of correct predictions.  Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.

```python
predictor_names = weekly.columns[1:-4].values
X = weekly[predictor_names]
```

```python
weekly['preds'] = log_reg_fit.predict(X)
weekly['pred_direction'] = 'Down'
weekly.loc[weekly['preds'] > 0.5, 'pred_direction'] = 'Up'
```

```python
log_reg_accuracy = accuracy_score(y_true=weekly['Direction'],
                            y_pred=weekly['pred_direction'])
log_reg_precision = precision_score(y_true=weekly['Direction'],
                            y_pred=weekly['pred_direction'], pos_label='Up')
log_reg_recall = recall_score(y_true=weekly['Direction'],
                            y_pred=weekly['pred_direction'], pos_label='Up')
print(F"Accuracy: {log_reg_accuracy:.4f}")
print(F"Precision: {log_reg_precision:.4f}")
print(F"Recall: {log_reg_recall:.4f}")
```

```python
conf_m = pd.DataFrame(confusion_matrix(y_true=weekly['Direction'],
                                       y_pred=weekly['pred_direction']),
                      columns = ['Down', 'Up'],
                      index = ['Down', 'Up'])
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

The confusion matrix and metrics above are telling us the following:

* TP = True Positive
* FP = False Positive
* TN = True Negative
* FN = False Negative

1) The accuracy is the percentage of correct predictions.  
    * $\text{Accuracy}=\frac{\text{correct predictions}}{\text{sample size}}=\frac{TN+TP}{TN+FP+FN+TP}=\frac{54 + 557}{54+48+430+557}=0.5611$
2) Precision tells us how many positive predictions were correct out of all the positive predictions that were made.
    * $\text{Precision}=\frac{\text{correct positive predictions}}{\text{all positive predictions}}=\frac{TP}{TP+FP}=\frac{557}{430+557}=0.5643$
3) Recall tells us how many of the true positives were correctly predicted.
    * $\text{Recall}=\frac{\text{correct positive predictions}}{\text{all positives}}=\frac{TP}{TP+FN}=\frac{557}{557+48}=0.9207$
    
Overall, this model is only making correct predictions about 56% of the time, not much better than choosing based on flipping a coin.  

Furthermore, the model has extremely high recall and is good about predicting when the market will go `Up`, but very rarely correctly predicts when the market will go `Down` as a consequence.  The model seems heavily biased towards predicting `Up` because out of 1089 predictions, 987 were `Up`.

### Part d) Logistic Regression using only `Lag2` as a predictor
Now fit the logistic regression model using a training data period from 1990-2008, with `Lag2` as the only predictor.  Compute the confusion matrix and the overall fraction of correct predictions for the held out data (that is, the data from 2009 and 2010).

After reducing our inputs down to just `Lag2`, the only statistically significant predictor, we achieve a testing accuracy of about 62.5%, which is a decent improvement over random guessing.  Additionally, 62.5% is a slight improvement over if one were to guess `Up` for every prediction, which would be correct about 59% of the time with our testing dataset.  

We continue to see high recall and that the vast majority of weeks that went `Up` were correctly classified.  One could potentially use this information to buy when the model predicts the market will increase next week.

##### Best subsets selection to determine which predictors yield the most optimal results

```python
def knn_tuner(training_x, training_y, testing_x, needs_scaling=False):
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.model_selection import GridSearchCV
    
    if needs_scaling == True:
        from scipy.stats import zscore
        training_x_scaled = zscore(training_x)
        testing_x_scaled = zscore(testing_x)
    
    
    knn_model = KNeighborsClassifier()
    param_grid = {'n_neighbors': np.arange(3,51,2)}
    knn_model_gscv = GridSearchCV(knn_model, 
                                  param_grid, 
                                  cv=5)
    knn_model_gscv.fit(training_x_scaled, training_y)
    
    optimal_n = knn_model_gscv.best_params_
    best_score = knn_model_gscv.best_score_
    preds = knn_model_gscv.predict(testing_x_scaled)
    
    return optimal_n, best_score, preds
```

```python
from tqdm import tnrange, tqdm_notebook
import itertools
```

```python
X = training_data.loc[:, 'Lag1':'Volume']
Y = training_data.loc[:, 'numerical_Direction']
x_test = testing_data.loc[:, 'Lag1':'Volume']
k = len(X.columns)

optimal_n_list, best_score_list, feature_list, preds_list = [], [], [], []
num_features = []

# Looping over k = 1 to k = 7 features in X
for k in tnrange(1, k + 1, desc = 'Loop...'):
    
    # Looping over all possible combinations: from 7 choose k
    for combo in itertools.combinations(X.columns, k):
        tmp_result = knn_tuner(X[list(combo)], Y, x_test[list(combo)], needs_scaling=True)
        optimal_n_list.append(tmp_result[0])
        best_score_list.append(tmp_result[1])
        feature_list.append(combo)
        preds_list.append(tmp_result[2])
        num_features.append(len(combo))
        
# Store in DataFrame
df = pd.DataFrame({'num_features': num_features,
                   'optimal_n': optimal_n_list,
                   'best_score': best_score_list,
                   'features': feature_list,
                   'preds': preds_list})
```

```python
knn_max_df = df[df.groupby('num_features')['best_score'].transform(max) == df['best_score']].sort_values('best_score', ascending=False)
knn_max_df
```

##### Best subsets selection

```python
def log_reg_tuner(training_data, testing_data, target_feature: str):
    import statsmodels.formula.api as smf
    from sklearn.metrics import accuracy_score, precision_score, recall_score   
    
    formula_string = target_feature + ' ~ '
    input_features = []
    for feature in training_data.columns:
        if feature != target_feature:
            input_features.append(feature)
    features_string = ' + '.join(input_features)
    formula_string += features_string
    #print(formula_string)
    log_reg_model = smf.logit(formula=formula_string, data=training_data)
    log_reg_fit = log_reg_model.fit(disp=0)
    
    preds = log_reg_fit.predict(testing_data[input_features])
    pred_w_labels = np.where(preds <= 0.5, 0, 1)
    
    
    log_reg_accuracy = accuracy_score(y_true=testing_data[target_feature],
                                y_pred=pred_w_labels)
    log_reg_precision = precision_score(y_true=testing_data[target_feature],
                                y_pred=pred_w_labels)
    log_reg_recall = recall_score(y_true=testing_data[target_feature],
                                y_pred=pred_w_labels)

    return log_reg_accuracy, log_reg_precision, log_reg_recall
```

Going forward, I plan to use `rad`, `nox`, `rm`, `age`, `tax`, and `zn` as my predictor variables.

```python
predictors = ['rad', 'nox', 'rm', 'age', 'tax', 'zn']

X_selected = X[predictors]
```

#### k-Fold Repeated Cross Validation (KFRCV)

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

lr_model_sk = LogisticRegression(max_iter=10000)


scores_all_features = cross_val_score(lr_model_sk, X_all, Y, scoring='accuracy', cv=cv, n_jobs=-1)

scores_selected_features = cross_val_score(lr_model_sk, X_selected, Y, scoring='accuracy', cv=cv, n_jobs=-1)
#lr_fit_all_vars = lr_model_sk.fit(X_selected, Y)
```

```python
print("k-Fold Repeated Cross Validation")
print("--------------------------------")
print(F"Average Testing Accuracy with all features: {scores_all_features.mean():.3f}")
print(F"Average Testing Error: {1 - scores_all_features.mean():.3f}")
print(F"Std Dev: {scores_all_features.std():.3f}")
print()
print(F"Average Testing Accuracy with selected features: {scores_selected_features.mean():.3f}")
print(F"Average Testing Error: {1 - scores_selected_features.mean():.3f}")
print(F"Std Dev: {scores_selected_features.std():.3f}")
```

Using KFRCV, we expect slightly better accuracy and roughly the same standard error, when using the features selected earier as opposed to used all features in the dataset.  Accuracy on testing data is estimated to be 85.4% with a standard error of 5.7%.

#### Predictions, Confusion Matrix, and Final Thoughts
